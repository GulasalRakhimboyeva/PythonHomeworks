```python
integer_list=[1,2,3,4,5,6,5,88]
even_count=0
odd_count=0
for char in integer_list:
    if char%2==0:
        even_count +=1
    else:
        odd_count +=1

print(even_count)
```

```python
list1=[1,2,3,4,5,6,5,88]
list2=[3,4,5]
if list2[::] in list1:
    print(list2, "is sublist of", list1)
else:
    print(list2, "is not sublist of", list1)
```
