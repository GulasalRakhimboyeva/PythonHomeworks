```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        seen={}
        for i, num in enumerate(nums):
            if target-num in seen:
                return [seen[target-num], i]
            seen[num]=i

sol=Solution()
sol.twoSum([3,5, 6], 9 )
```
