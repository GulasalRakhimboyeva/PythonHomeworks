**Step 4: Email va phone validation **
   - Emaillarni kichik harflarga o‘tkazish va noto‘g‘ri formatdagi emaillarni aniqlash. 
   - Telefon raqamlarni tozalash va yagona standart formatga keltirish (masalan: +998...). 

```python
df['email'].head(30)
```

```python
# emaillarni kichik harfga o`tkazdik`
df['email']=df['email'].str.lower()

email_pattern = r'^[a-z0-9._%+-]+@[a-z0-9.-]+\.[a-z]{2,}$'
df['email_valid']=df['email'].str.match(email_pattern, na=False)
df['email_valid'].sum()
```

```python
# invalid emaillarni nanga o`zgartirish`
df.loc[~df['email_valid'], 'email']=np.nan

df = df.drop('email_valid', axis=1)
df['email'].head(50)
```

```python
# phone number tozalash

def standardize_phone(x):
    if pd.isna(x):
        return np.nan
    x=str(x)
    x=x.split('x')[0]
    x=''.join(filter(str.isdigit, x))

    if x.startswith('1') and len(x)==11:
        return '+'+x
    elif x.startswith('001') and len(x)>11:
        return '+' + x[2:]
    elif len(x)==10:
        return '+1'+x
    else:
        return np.nan
df['phone_clean']=df['phone'].apply(standardize_phone)
df['phone']=df['phone_clean']
df.drop('phone_clean', axis=1, inplace=True)
df['phone'].head(50)
```

**Step 5: JSON parsing**
   - JSON columnlarini ochish (profile_json). 
   - JSON ichidagi ma’lumotlarni alohida columnlarga ajratish: hobbies, skills, family, devices. 
   - JSON ichidagi array yoki nested objectlarni tekshirish va flatten qilish.

```python
import ast 
import json
def json_parse(x):
    if pd.isna(x) or x=='':
        return {}
    try:
        return json.loads(x)
    except json.JSONDecodeError:
        try:
            return ast.literal_eval(x)
        except:
            return {}
df['profile_json_parsed'] = df['profile_json'].apply(json_parse)

```

```python

import copy
df1=copy.deepcopy(df)
```

```python
df1_flat = pd.json_normalize(
    df1['profile_json_parsed'],
    sep='_'
)
df1_flat
```

```python
df1_flat['hobbies']=df1_flat['hobbies'].apply(lambda x: ', '.join(x) if isinstance(x, list) else np.nan)
df1_flat['skills_soft']=df1_flat['skills_soft'].apply(lambda x: ', '.join(x) if isinstance(x, list) else np.nan)
```

```python
import numpy as np

def extract_devices(devices):
    result = {
        'laptop_brand': np.nan,
        'laptop_year': np.nan,
        'phone_brand': np.nan,
        'phone_year': np.nan
    }
    
    if isinstance(devices, list):
        for d in devices:
            if d.get('type') == 'laptop':
                result['laptop_brand'] = d.get('brand')
                result['laptop_year'] = d.get('year')
            elif d.get('type') == 'phone':
                result['phone_brand'] = d.get('brand')
                result['phone_year'] = d.get('year')
                
    return pd.Series(result)

df1_flat[['laptop_brand','laptop_year','phone_brand','phone_year']] = \
    df1_flat['devices'].apply(extract_devices)
```
